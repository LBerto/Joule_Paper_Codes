```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.colors as colors
from matplotlib import ticker, cm
import matplotlib.image as mpimg
from mpl_toolkits.axes_grid1 import AxesGrid
from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)
%matplotlib inline
```

```python
BD_1E3_1E17 = np.loadtxt('BD_S_1E3_1E17.txt')
BD_1E5_1E17 = np.loadtxt('BD_S_1E5_1E17.txt')
BD_1E3_1E15 = np.loadtxt('BD_S_1E3_1E15.txt')
BD_1E5_1E15 = np.loadtxt('BD_S_1E5_1E15.txt')
```

```python
x = BD_1E5_1E17[0,]*1E7
Ec1 = BD_1E3_1E17[1,]
Ev1 = BD_1E3_1E17[2,]
Efn1 = BD_1E3_1E17[3,]
Efp1 = BD_1E3_1E17[4,]

Ec2 = BD_1E5_1E17[1,]
Ev2 = BD_1E5_1E17[2,]
Efn2 = BD_1E5_1E17[3,]
Efp2 = BD_1E5_1E17[4,]

Ec3 = BD_1E3_1E15[1,]
Ev3 = BD_1E3_1E15[2,]
Efn3 = BD_1E3_1E15[3,]
Efp3 = BD_1E3_1E15[4,]

Ec4 = BD_1E5_1E15[1,]
Ev4 = BD_1E5_1E15[2,]
Efn4 = BD_1E5_1E15[3,]
Efp4 = BD_1E5_1E15[4,]
```

```python
minorLocator_x = MultipleLocator(10)
majorLocator_x = MultipleLocator(100)
minorLocator_y = MultipleLocator(0.1)
majorLocator_y = MultipleLocator(0.5)

fig = plt.figure(num=None, dpi=200, facecolor='w', edgecolor='k')
ax1 = fig.add_subplot(111)
plt.rcParams['axes.linewidth'] = 1.5
ax1.xaxis.set_minor_locator(minorLocator_x)
ax1.xaxis.set_major_locator(majorLocator_x)
ax1.yaxis.set_minor_locator(minorLocator_y)
ax1.yaxis.set_major_locator(majorLocator_y)
ax1.set_xlim(0, 800)
ax1.set_ylim(0,2)
ax1.plot(x, Ec1, 'k', linewidth=2)
ax1.plot(x, Ev1, 'k', linewidth=2)
ax1.plot(x, Efn1, '--b', linewidth=2)
ax1.plot(x, Efp1, '--r', linewidth=2)
ax1.tick_params(axis='both', which='major', labelsize=15)
# get handles
handles, labels = ax1.get_legend_handles_labels()
# use them in the legend
ax1.legend(handles, labels, loc='lower right',numpoints=1, frameon=False,fontsize=10)
plt.show()
fig.savefig('Figure4A.TIFF',bbox_inches='tight')
```
